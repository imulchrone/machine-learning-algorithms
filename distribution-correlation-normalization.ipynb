There are 13 columns and 178 rows in the dataset. Every attribute is a float without any null values.

```python
#Histograms
fig, ax = plt.subplots(3,1, figsize = (6,15))

#Alcohol
ax[0].hist(wine["alcohol"],bins=10)
ax[0].set_title("Alcohol")

#Malic acid
ax[1].hist(wine["malic_acid"], bins=10)
ax[1].set_title("Malic Acid")

#Ash
ax[2].hist(wine["ash"], bins=10)
ax[2].set_title("Ash")

plt.show()
```

Alcohol and ash both have fairly normal distributions. There may be slight concerns of bimodality in alchohol but the overall distrubtion is bell shaped. Malic acid has a right skew, with the mode being around 1.5 and the maximum being over 5.

```python
#Boxplots
fig, ax = plt.subplots(1,3, figsize = (10,6))

#Alcohol
ax[0].boxplot(wine["alcohol"])
ax[0].set_title("Alcohol")

#Malic acid
ax[1].boxplot(wine["malic_acid"])
ax[1].set_title("Malic Acid")

#Ash
ax[2].boxplot(wine["ash"])
ax[2].set_title("Ash")

plt.show()
```

There are no outliers in alcohol and the boxplot supports the normal distrbution displayed by the histogram. With the right skew of malic acid, there are a few outliers at the upper end of the range. Ash has outliers at both the minimum and maximum ends with its IQR being the smallest of the three, condesning its distribution closer to the median of about 2.35.

```python
corr_matrix = wine.corr()
corr_matrix
```

```python
corr_matrix["alcohol"].sort_values(ascending=False)
```

Proline and color intensity both have moderate positive correlations with alcohol level. Alcalinity of ash has the strongest negative relationship with alcohol, but the correlation is fairly low. We can plot these relationships to see the patterns.

```python
#Scatterplots
fig, ax = plt.subplots(3,1, figsize = (6,15))

#Proline
ax[0].scatter(wine["proline"],wine["alcohol"])
ax[0].set_xlabel("Proline")
ax[0].set_ylabel("Alcohol")

#Color intensity
ax[1].scatter(wine["color_intensity"],wine["alcohol"])
ax[1].set_xlabel("Color Intensity")
ax[1].set_ylabel("Alcohol")

#Alcalinity of ash
ax[2].scatter(wine["alcalinity_of_ash"],wine["alcohol"])
ax[2].set_xlabel("Alcalinity of Ash")
ax[2].set_ylabel("Alcohol")

plt.show()
```

The scatterplots clearly show the relationships between alcohol and the three variables. Proline shows a stronger positive linear realtionship, whereas the cone shape of the color intensity plot suggests that when color intensity increases it may not have as strong an effect on alcohol. Alcalinity of ash does not display a strong relationship which was expected given its low correlation value, but there is a negative trend visible in the plot.

```python
#Scatterplot flavanoids vs total_phenols
plt.scatter(wine["flavanoids"],wine["total_phenols"])
plt.xlabel("Flavanoids")
plt.ylabel("Total Phenols")
plt.annotate("corr = %0.4f"%corr_matrix["total_phenols"].loc["flavanoids"],
             xy=(4,1),
             xytext=(4,1),
             bbox = dict(boxstyle ="round", fc ="0.8"))
plt.show()
```

The strongest correlation in the dataset is between flavanoids and total phenols. With a value of 0.8646, the high correlation suggests a strong linear relationship between the two variables.

The values of magnesium and proline are signifcantly larger than the rest of the variables. We can scale these down using min-max normalization. Min-max is appropriate in this situation since there are no negative values in these variables.

```python
wine_norm = wine.copy()
wine_norm["magnesium_minmax"] = ((wine_norm["magnesium"] - wine_norm["magnesium"].min()) / 
                            (wine_norm["magnesium"].max() - wine_norm["magnesium"].min()))
wine_norm["proline_minmax"] = ((wine_norm["proline"] - wine_norm["proline"].min()) / 
                            (wine_norm["proline"].max() - wine_norm["proline"].min()))
```
